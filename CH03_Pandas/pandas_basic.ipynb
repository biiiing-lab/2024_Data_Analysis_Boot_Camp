```python
dust = pd.read_excel('./file/dust1.xlsx')
dust.head()
```

```python
dust = dust.drop(['지역', '망', '측정소코드'], axis=1)
dust.head()
```

```python
dust['city'] = '서울'
dust.head()
```

```python
dust.rename(columns={
    '측정소명' : 'name',
    '측정일시' : 'date',
    '주소' : 'addr'
}, inplace=True)

dust.columns
```

```python
dust.dtypes
```

```python
dust['date'] = dust['date'].astype(str)
dust['date'] = dust['date'].str.slice(0, 8)
dust.dtypes
```

```python
dust['date'] = pd.to_datetime(dust['date'])
dust.head()
```

```python
dust['year'] = dust['date'].dt.year
dust['month'] = dust['date'].dt.month
dust['day'] = dust['date'].dt.day
dust.head()
```

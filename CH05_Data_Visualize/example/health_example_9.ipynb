```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('../file/health_screenings_2020_1000ea.xlsx')
data
```

```python
data9 = data.loc[:, ['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]

data9.loc[data9['gender']==1, ['gender']] = 'male'
data9.loc[data9['gender']==2, ['gender']] = 'female'
data9.loc[(data9['smoking']==1)| (data9['smoking']==2), ['smoking']] = 'non-smoking'
data9.loc[data9['smoking']==3, ['smoking']] = 'smoking'

smoking = data9.groupby(['gender', 'smoking'])['smoking'].count()
smoking = smoking.to_frame(name = 'count') # 데이터 프레임화
smoking = smoking.reset_index()
```

```python
plt.figure(figsize=(10,6))

# 답 (카운트플롯 : 카테고리별로 데이터 개수 세주는 그래프)
sns.countplot(data=data9, x='smoking', hue='gender', order=['smoking', 'non-smoking'])
plt.title('smoking count plot graph')
plt.show()
```

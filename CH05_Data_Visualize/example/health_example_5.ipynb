```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('../file/health_screenings_2020_1000ea.xlsx')
data
```

```python
data5 = data.loc[:, ['city_code', 'gender', 'waist']]
waistData = data5.loc[100:110, ['waist']] # 11건
```

```python
plt.figure(figsize=(10, 5))

# 100~110번까찌의 인덱스를 x 축으로 사용
plt.bar(waistData.index, waistData['waist'], color = 'c')

plt.ylim(50, 110)
plt.xlabel('count')
plt.ylabel('waist')
plt.title('2020 health screenings waist')
plt.grid()
plt.show()
```

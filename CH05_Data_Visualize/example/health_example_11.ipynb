```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('../file/health_screenings_2020_1000ea.xlsx')
data
```

```python
data11 = data.loc[:, ['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
```

```python
plt.figure(figsize=(10,6))

correlation_data11 = data11.corr()
upp_mat = np.triu(correlation_data11) # ì˜¤ë¥¸ìª½ ìœ—ë¶€ë¶„

# ë‹µ 
sns.heatmap(correlation_data11, annot=True, cmap='YlGnBu', mask=upp_mat)

plt.title('heat map graph')
plt.show()
```

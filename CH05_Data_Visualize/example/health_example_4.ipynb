```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('../file/health_screenings_2020_1000ea.xlsx')
data
```

```python
# 도시별 그룹
data4 = data.loc[:, ['city_code', 'gender', 'waist']]

male_data = data4[data4['gender']==1].groupby('city_code')['waist'].mean()
female_data = data4[data4['gender']==2].groupby('city_code')['waist'].mean()
```

```python
plt.figure(figsize=(12, 4))

# 빈칸
plt.plot(male_data, 'bo--', label='male')
plt.plot(female_data, 'ro--', label='female')

plt.xlabel('count')
plt.ylabel('waist')
plt.title('2020 health screenings waist')
plt.legend()
plt.grid()
plt.show()
```

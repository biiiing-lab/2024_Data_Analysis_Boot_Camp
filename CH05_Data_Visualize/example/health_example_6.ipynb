```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('../file/health_screenings_2020_1000ea.xlsx')
data
```

```python
lEyeData = data['eye_left']
rEyeData = data['eye_right']
```

```python
plt.figure(figsize=(5, 5))

# 답
plt.scatter(lEyeData, rEyeData)

plt.xlim(0, 2.5)
plt.ylim(0, 2.5)
plt.xlabel('eye left')
plt.ylabel('eye right')
plt.title('2020 health screenings scatter')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel('../file/health_screenings_2020_1000ea.xlsx')
data
```

```python
data7 = data.loc[:, ['city_code', 'gender', 'waist']]

male_data = data7.loc[data7.gender==1, ['gender','waist']]
female_data = data7.loc[data7.gender==2, ['gender','waist']]
```

```python
plt.figure(figsize=(10, 6))

# ë‹µ
plt.hist(male_data['waist'], bins=8, alpha=0.5, label='male' )
plt.hist(female_data['waist'], bins=8, alpha=0.5, label='female' )

plt.xlim(50, 115)
plt.xlabel('waist')
plt.ylabel('frequency')
plt.legend()
plt.grid()
plt.show()
```

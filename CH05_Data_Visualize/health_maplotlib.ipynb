```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('./file/health_screenings_2020_1000ea.xlsx')
```

```python
# 키 및 몸무게 산점도 그래프
heightdata = data['height']
weightdata = data['weight']

plt.figure(figsize=(10, 4))
plt.scatter(heightdata, weightdata)

plt.xlabel('height')
plt.ylabel('weight')

plt.title('2020 health screeings scatter graph')
plt.grid()
plt.show()
```

```python
# 수축기, 이완기 혈압, 공복 혈당 그룹 산점도 그래프

systolic_data = data['systolic']
diatolic_data = data['diastolic']
blood_sugar_data = data['blood_sugar']

plt.figure(figsize=(10, 6))
plt.scatter(systolic_data, diatolic_data, color='r', edgecolors='w', label='systolic * diatolic')
plt.scatter(systolic_data, blood_sugar_data, color='g', edgecolors='w', label='systolic * blood_sugar')
plt.scatter(diatolic_data, blood_sugar_data, color='b', edgecolors='w', label='diatolic * blood_sugar')

plt.xlim(40, 180)
plt.ylim(30, 300)
plt.title('2020 health screenings group scatter graph')
plt.legend()
plt.show()
```

```python
# 어떤 구간에 어떤 데이터가 얼마나 있는가 : 히스토그램 사용
# 성별에 따른 키 히스토그램

female_data = data.loc[data.gender==2, ['gender', 'height']]

plt.figure(figsize=(10, 6))
plt.hist(female_data['height'], bins=7)

plt.xlim(130, 180)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 health screenings female height histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 키 그룹 히스토그램
female_data = data.loc[data.gender==2, ['gender', 'height']]
male_data = data.loc[data.gender==1, ['gender', 'height']]

plt.figure(figsize=(10, 6))
plt.hist(female_data['height'], bins=7, alpha=0.5, label='female')
plt.hist(male_data['height'], bins=7, alpha=0.5, label='male')

plt.xlim(130, 195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 health screenings female height histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 키 그룹 누적 히스토그램
height = np.array([male_data['height'], female_data['height']], dtype=object)

plt.figure(figsize=(10, 6))
plt.hist(height, bins=6, label=['male', 'female'], stacked=True)

plt.xlim(130, 195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 health screenings female height histogram')
plt.legend()
plt.grid()
plt.show()
```

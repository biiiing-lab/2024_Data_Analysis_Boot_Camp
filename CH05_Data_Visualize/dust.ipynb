```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('./file/fine_dust.xlsx', index_col='area')
data2018 = data[2018]
```

```python
# 2018    
plt.figure(figsize=(15, 4))
plt.plot(data2018, color='b', marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 fine dust line graph')
plt.grid()
plt.show()
```

```python
# 2016 ~ 2019
plt.figure(figsize=(15, 4))
for year in range(2016, 2020) :
    chardata = data[year]
    plt.plot(chardata, marker='s', label='year')

plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016~2019 fine dust line graph')
plt.grid()
plt.show()
```

```python
data2017 = data[2017]

plt.figure(figsize=(15, 4))
plt.bar(data2017.index, data2018, color='g')
plt.ylim(35, 55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2017 fine dust line graph')
plt.grid()
plt.show()
```

```python
index = np.arange(4)
plt.figure(figsize=(15, 4))

data4 = data.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016, 2020) :
    chardata = data4[year]
    plt.bar(index, chardata, width=0.2, label=year)
    index = index + 0.2

plt.ylim(35, 55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.xticks(index-0.5, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])
plt.title('2016~2019 fine dust line graph')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
index = np.arange(4)


data4 = data.loc['Gyeonggi':'Daegu', 2016:2019]
for year in range(2016, 2020) :
    chardata = data4[year]
    plt.barh(index, chardata, label=year)


plt.xlim(30, 55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.yticks(index, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])
plt.title('2016~2019 fine dust group barh graph')
plt.legend()
plt.show()
```

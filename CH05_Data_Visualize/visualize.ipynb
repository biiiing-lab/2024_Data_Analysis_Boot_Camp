```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2, 3, 4, 5]
y = [1, 3, 5, 8]
```

```python
plt.title('Line graph')
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.legend() # 범례
```

```python
plt.plot(x, y, color = 'red')
plt.xlabel('x value')
plt.ylabel('y value')
```

```python
plt.plot(x, color='r', label='--', linestyle='--')
plt.plot(y, color='g', label='..', linestyle=':')
plt.show()
```

```python
x = [5, 10, 15, 20]
y = [1, 3, 5 ,7]
```

```python
plt.plot(x, y, color='blue', linestyle = '--', marker ='o', markersize='10')
# marker 종류 : o, s, *, +, . 등
plt.xlim(0, 25)
plt.ylim(-2, 10)
plt.show()
```
